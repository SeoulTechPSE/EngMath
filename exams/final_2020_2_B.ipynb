**1.** $\text{ }$ Find the eigenvalues and eigenfunctions of the problem:

> $\begin{align*}
\left [ (1 + x)^2 u' \right ]' +\lambda u &= 0,\;\; 0 < x < 1\\ 
 u(0) = u(1) &= 0 
\end{align*}$

>**HINT** $\;\;(1 +x)^{ia} = e^{ia \ln (1 +x)}$

**solution**

>$\begin{align*}
 &\Downarrow 1 +x = t \\ 
 t^2 \frac{d^2 u}{dt^2} &+2t \frac{du}{dt} +\lambda u = 0, \;\;\;\text{Euler equation} \\ 
 &\Downarrow u = t^r \\ 
 t^r &\left[ r(r -1) +2r +\lambda \right] = 0\\ 
 &\Downarrow \\
 r &= -\frac{1}{2} \pm i \frac{1}{2} \sqrt{ 4\lambda -1}, \;\; \lambda > \frac{1}{4} \\
 &\Downarrow \\
 u(x) &= (1 +x)^{-\frac{1}{2}} \left[ c_1 \cos \left( \frac{1}{2} \sqrt{4\lambda -1} \ln(1 +x) \right) 
      +c_2 \sin \left( \frac{1}{2} \sqrt{4\lambda -1} \ln(1 +x) \right) \right] \\
 &\Downarrow  u(0) = u(1) = 0 \\ 
 0 &= c_1\\ 
 0 &= \frac{1}{\sqrt{2}} \, c_2 \, \sin \left(\frac{\ln2}{2} \sqrt{4\lambda -1} \right ) \;\; \Rightarrow \;\; \frac{\ln2}{2} \sqrt{4\lambda -1} = n\pi, \;\;n=1,2,3,\cdots \\
 &\Downarrow \\
 \text{Eigenvalues:}& \;\; \lambda_n = \left( \frac{n\pi}{\ln2} \right)^2 + \frac{1}{4}, \;\; n= 1,2,3,\cdots\\
 \text{Eigenfunctions:}& \;\; u_n(x) = (1 +x)^{-\frac{1}{2}} \sin \left[ \frac{n\pi}{\ln2} \ln(1 +x)\right]
\end{align*}$

**2.** $\text{ }$ Solve

>$\begin{align*}
 u_t &= k \,u_{xx}, \;\; a < x < b, \;\; t > 0 \\ 
 &u(a, t) = 0 \\ 
 &u(b, t) = 0 \\ 
 &u(x, 0) = (x -a)(b -x)
\end{align*}$

**solution** $\text{ }$ 

> $\begin{align*}
 &\Downarrow \;\; \hat{x} = x -a, \;\;\hat{b}= b -a\\
 \\
 u_t &= k \,u_{\hat{x}\hat{x}}, \;\; 0 < \hat{x} < \hat{b}, \;\; t > 0 \\ 
 &u(0, t) = 0 \\ 
 &u(\hat{b}, t) = 0 \\ 
 &u(\hat{x}, 0) = \hat{x}(\hat{b} -x)\\
 \\
 &\Downarrow u(\hat{x}, t) = X(\hat{x})T(t)\\
 \\
 u(\hat{x}, t) &= \sum_{n=1}^{\infty}c_n \exp \left[ -k\left( \frac{n\pi}{\hat{b}}\right)^2 t  \right] \sin \frac{n\pi\hat{x}}{\hat{b}} \\
 \\
 &\Downarrow t=0\\
 \\
 \hat{x}(\hat{b} -\hat{x}) &= \sum_{n=1}^{\infty}c_n \sin \frac{n\pi\hat{x}}{\hat{b}} \;\;\Rightarrow\;\;
 c_n = \frac{2}{\hat{b}} \int_0^{\hat{b}} \,\hat{x}(\hat{b} -\hat{x}) \, \sin \frac{n\pi\hat{x}}{\hat{b}}\,d\hat{x}
 =4\frac{\hat{b}^2}{(n\pi)^3} \left[ 1-(-1)^n \right]\\
 \\
 &\Downarrow\\
 \\
 u(x, t) &= \frac{8(b -a)^2}{\pi^3} \sum_{n=1}^{\infty}\frac{1}{(2n -1)^3} \exp \left[ -k \left( \frac{(2n -1)\pi}{b -a}\right)^2 t  \right] \sin \frac{(2n -1)\pi(x-a)}{b -a}
 \end{align*}$

**3.** $\text{ }$ Solve by separation of variables

>$\begin{align*}
 u_{tt} +u_t &= u_{xx},\;\;\;0 < x < \pi, \;\; t > 0 \\ 
 u(0, t)&= 0\\ 
 u_x(\pi, t)&= 0\\ 
 u(x, 0)&= \sin^3 \frac{x}{2}\\ 
 u_t(x, 0)&= 0 
\end{align*}$

**solution**

>$\begin{align*}
 &\Downarrow \;\;  u(x,t) = X(x) T(t) \\
 \\
 \frac{T'' +T'}{T} &= \frac{X''}{X}=-\lambda < 0 \\
 \\
 &\Downarrow \\ 
 \\
 X'' &+\lambda X = 0, \;\;X(0)=X'(\pi)=0 \\
 T'' &+T'+\lambda T = 0 \\
 \\
 &\Downarrow \\ 
 \\
 \lambda_n &= \left( \frac{2n-1}{2} \right)^2, \;\;n=1,2,3,\cdots \\
 X_n &= \sin \frac{2n-1}{2}x \\
 T_1 &= e^{-t/2} \left( a_1 +b_1 t \right) \\
 T_n &= e^{-t/2} \left[ a_n \cos \left( \sqrt{n(n -1)}\, t\right) +b_n \sin \left(\sqrt{n(n -1)}\, t\right) \right], \;\; n =2,3,\cdots \\
 \\
 &\Downarrow \\ 
 \\
 u(x,t) &= e^{-t/2} \left[ (a_1 +b_1 t) \sin\frac{x}{2} +\sum_{n=2}^\infty \left(a_n \cos \sqrt{n(n-1)}\, t +b_n \sin \sqrt{n(n-1)}\, t \right ) \,\sin \frac{2n-1}{2}x\right ] \\
 u_t(x,t) &= -\frac{1}{2} u(x,t) \;+\; e^{-t/2} \left[ b_1 \sin\frac{x}{2} +\sum_{n=2}^\infty \left(-a_n \sqrt{n(n-1)}\sin \sqrt{n(n-1)}\, t +b_n \sqrt{n(n-1)} \cos \sqrt{n(n-1)}\, t \right ) \,\sin \frac{2n-1}{2}x\right]
 \\ \\
 &\Downarrow\;\;t=0 \\ 
 \\
 u(x,0) &= \sin^3\frac{x}{2}=\frac{3}{4} \sin \frac{1}{2}x -\frac{1}{4} \sin \frac{3}{2}x = a_1 \sin\frac{1}{2}x +\sum_{n=2}^\infty a_n \sin \frac{2n -1}{2} x \;\;\Rightarrow \;\; a_1 =\frac{3}{4}, \;\;a_2 =-\frac{1}{4} \\
 u_t(x,0) &= 0 =-\frac{3}{8} \sin\frac{1}{2}x +\frac{1}{8} \sin \frac{3}{2}x +b_1 \sin\frac{1}{2}x +\sum_{n=2}^\infty b_n \sqrt{n(n-1)} \sin\frac{2n-1}{2}x \;\;\Rightarrow \;\; b_1 =\frac{3}{8}, \;\;b_2 =-\frac{\sqrt{2}}{16} \\
 \\
 &\Downarrow \\ 
 \\
 u(x,t) &= \frac{1}{4} e^{-t/2} \left[ 3\left(1 +\frac{1}{2}t\right) \sin \frac{1}{2}x -\left( \cos \sqrt{2}t +\frac{\sqrt{2}}{4} \sin \sqrt{2}t\right) \sin\frac{3}{2}x  \right]
\end{align*}$

**4.** $\text{ }$ Solve 

>$\begin{align*}
 \nabla^2 u &= 0, \;\;\;\;\;\;\;\;\;\; r < 1 \\ 
 u_r &= g(\phi, \theta), \;\; r=1 
\end{align*}$

where

>$\displaystyle\int_0^{2\pi} \int_0^\pi g(\phi,\theta) \, \sin\phi \,d\phi \,d\theta = 0$

Let $u = 0$ at $r = 0$. What happens if this integral is not zero?  


**solution**


>$\begin{align*}
    \nabla^2 u &=u_{rr} +\frac{2}{r} u_r +\frac{1}{r^2}u_{\phi\phi} +\frac{\cot\phi}{r^2} u_{\phi} +\frac{1}{r^2\sin^2\phi}u_{\theta\theta} =0\\ \\
 &\Downarrow \;\;  u(r,\phi,\theta) = R(r) \Phi(\phi) \Theta(\theta) \\
 \\
 - \frac{r^2R''+2rR'}{R} \sin^2 \phi &- \frac{\left[ \sin\phi\, \Phi'\right ]'}{\Phi} \sin\phi=\frac{\Theta''}{\Theta} = -\mu < 0\\
 - \frac{r^2R''+2rR'}{R} &= \frac{\left[ \sin\phi\, \Phi'\right ]'}{\sin\phi\,\Phi} -\frac{\mu}{\sin^2\phi} = -\lambda < 0 \\
 \\ \Downarrow \;\;t=\cos\phi\\ \\
 \Theta'' +\mu \Theta = 0,& \;\; \Theta(0)=\Theta(2\pi), \;\;\Theta'(0)=\Theta'(2\pi) \;\;\Rightarrow \;\; \mu_m = m^2, \;\;m=0,1,2,\cdots, \;\;\Theta_m(\theta) =\cos m\theta, \;\sin m\theta \\
 \frac{d}{dt} \left[ (1 -t^2) \frac{d\Phi}{dt}\right] &- \frac{m^2}{1-t^2} \Phi +\lambda \Phi = 0, \;\;\lambda_n =n(n+1), \;\;n=0,1,2,\cdots,\;\;\Phi_{mn}(t)=P_n^m(t) = (1 -t^2)^{m/2} \frac{d^m}{dt^m}P_n(t)\\
 r^2 R_n'' +2r R_n'&-n(n+1)R_n = 0, \;\; R_n = r^n\; (\because u \text{ is bounded}) \\ \\
 \Downarrow \\
 u(r,\phi,\theta) = \frac{1}{2} a_{00} &+\sum_{n=1}^\infty r^n \left[ \frac{1}{2}a_{n0} P_n(\cos\phi) +\sum_{m=1}^n 
  \left( a_{nm} \cos m\theta +b_{nm} \sin m\theta \right) P_n^m(\cos\phi)\right] \\ \\
   &\Downarrow \;\;u=0 \;\text{ at }\; r=0\;\Rightarrow\; a_{00}=0, \;\;u_r(1,\phi,\theta)=g(\phi,\theta)\\ \\
   u_r(1,\phi,\theta) &= g(\phi,\theta)=\sum_{n=1}^\infty n \left[ \frac{1}{2}a_{n0} P_n(\cos\phi) +\sum_{m=1}^n 
  \left( a_{nm} \cos m\theta +b_{nm} \sin m\theta \right) P_n^m(\cos\phi)\right] \\
  a_{nm}&=\frac{1}{n\pi}\frac{(2n +1)}{2}\frac{(n -m)!}{(n+m)!} \,\int_0^{2\pi} \int_0^\pi g(\phi,\theta) \cos m\theta  \,P_n^m(\cos\phi)\,\sin\phi \,d\phi \,d\theta \\
  b_{nm}&=\frac{1}{n\pi}\frac{(2n +1)}{2}\frac{(n -m)!}{(n+m)!} \,\int_0^{2\pi} \int_0^\pi g(\phi,\theta) \sin m\theta  \,P_n^m(\cos\phi)\,\sin\phi \,d\phi \,d\theta \\
\end{align*}$

> By the divergence theorem

> $\;\;\;\displaystyle\int_0^{2\pi} \int_0^\pi \int_0^1 (\nabla^2 u) \,r^2\sin \phi \,dr\,d\phi\,d\theta = \int_0^{2\pi} \int_0^\pi \frac{\partial u}{\partial r}(1,\phi,\theta) \,\sin\phi \,d\phi\,d\theta$

> Therefore if the integral on the right is not zero, $u$ cannot satisfy Laplace equation

**1[20].** $~$ Find the solution of

> $
\begin{align*}
 u_t &= u_{xx} + \sin \lambda_1 x, \;\;0<x<1, \;\; 0<t<\infty\\ 
 &u(0,t)=0\\ 
 &u_x(1,t) + u(1,t)=0 \\ 
 &u(x,0)=0 
\end{align*}
$

> where $\lambda_1$ is the first root of the equation $\tan \lambda = -\lambda$

> **Solution**

\begin{align*}
 w_t &= w_{xx}\\ 
 &w(0,t)=0 \\ 
 &w_x(1,t)+w(1,t)=0 \\ 
 &w(x,0)=0 \\
 &\Downarrow {\scriptsize w(x,t)=T(t)X(x)} \\
 X''&+\lambda^2 X=0\\ 
    &X(0)=0\\
    &X'(1)+X(1)=0\\
 &\Downarrow \\
 X &= c_1 \cos\lambda x + c_2 \sin\lambda x \\
 X' &=-c_1\lambda\sin\lambda x +c_2\lambda \cos\lambda x \\
 &\Downarrow \\
 X(0) &= c_1 = 0 \\
 X'(1) &+X(1)=c_2(\lambda\cos\lambda+\sin\lambda) = 0 \;\rightarrow \;\tan\lambda=-\lambda, \;c_2 \neq 0 \\
  &\Downarrow \\
  \lambda_n&, \;\;n=1,2,3,\cdots \\
  X_n & = \sin \lambda_n x
\end{align*}

> **Fourier Transform**

\begin{align*}
 \mathcal{F}[u(x,t)] &= 2\int_0^1 u(x,t) \sin \lambda_n x \,dx = U_n(t) \\ 
 \mathcal{F}^{-1}[U_n(t)] &= \sum_{n=1}^\infty U_n(t) \sin\lambda_n x \\ 
 &\Downarrow \\ 
 \frac{d U_n}{dt} &+\lambda_n U_n = 
   \begin{cases}
     \;1, & \text{when } n=1 \\ 
     \;0, & \text{when } n\neq 1 
   \end{cases} \\ 
 & U_n(0) = 0 \\
 &\Downarrow \\
 U_1(t) &=\frac{1}{\lambda_1} \left( 1-e^{-\lambda_1 t} \right) \\
 U_n(t) &= 0 \\
  &\Downarrow \\
 u(x,t) &= \frac{1}{\lambda_1} \left( 1-e^{-\lambda_1 t} \right) \sin\lambda_1 x
\end{align*}

**2[20].** $~$ Solve the semi-infinite string problem

> $
\begin{align*}
 u_{tt} &= u_{xx}, \;\;0<x<\infty, \; 0 < t < \infty\\ 
 u(0, t) &= 0\\
 u(x, 0) &=xe^{-x^2} \\
 u_t(x, 0) &= 0
\end{align*}
$

> Draw the solution for various values of time

> **Solution**

\begin{align*}
 u(x, 0) &= x e^{-x^2}=f(x) \\
 u_t(x, 0) &= 0 =g(x) \\
 &\Downarrow \\
 u(x,t) &= 
   \begin{cases}
     \frac{1}{2} \left[ f(x+t) + f(x-t) \right], & x \geq t \\ 
     \frac{1}{2} \left[ f(x+t) - f(t-x) \right], & x < t
    \end{cases}\\ 
   &= 
\frac{1}{2} \left[ (x+t) e^{-(x+t)^2} + (x-t) e^{-(x-t)^2} \right]
\end{align*}

**3[20].** $~$ Solve the damped vibrating-string problem

> $
\begin{align*}
 u_{tt}=c^2 &u_{xx} -\beta u_t, \;\;0<x<1, \;\;0<t<\infty \\ 
 u(0,t)&=0 \\ 
 u(1,t)&=0 \\ 
 u(x,0)&=f(x) \\ 
 u_t(x,0)&=0 
\end{align*}
$

> **Solution**

\begin{align*}
 u(x,t) &= T(t)X(x)\\ 
 &\Downarrow \\ 
 \frac{T''+\beta T'}{c^2 T} &= \frac{\,X''}{X} = -\lambda < 0 \\ 
 &\Downarrow \\ \\
 1) \;\; &X''+\lambda X=0, \;\;X(0)=X(1)=0 \\
  &\lambda_n = (n\pi)^2, \;\; n=1,2,\cdots \\
  &X_n = \sin n\pi x \\ \\
 2) \;\; &T_n'' +\beta T_n' +(cn\pi)^2 T_n = 0 \\
  & {\scriptsize p_n^\pm = \frac{-\beta \pm \sqrt{\beta^2 - (2cn\pi)^2}}{2}} \\
  &T_n=c_{n,1}\, e^{p_n^+ t} + c_{n,2}\, e^{p_n^- t} \\ \\
 &\Downarrow \\ 
 u(x,t) &= \sum_{n=1}^\infty \left( c_{n,1}\, e^{p_n^+ t} + c_{n,2}\, e^{p_n^- t} \right) \sin n\pi x \\
 &\Downarrow \text{ at } \;t= 0\\
 f(x) &= \sum_{n=1}^\infty \left(c_{n,1} +c_{n,2} \right ) \sin n\pi x \\
 0 &= \sum_{n=1}^\infty \left(c_{n,1} p_n^+ +c_{n,2}p_n^- \right ) \sin n\pi x \\
 &\Downarrow \\
 2\int_0^1 f(x) \sin n\pi x\,dx &= c_{n,1} +c_{n,2} \\
 0 &= c_{n,1}\,p_n^+ +c_{n,2}\,p_n^- \\
 &\Downarrow \\
 \begin{bmatrix}
   c_{n,1} \\ c_{n,2} 
 \end{bmatrix} &= \frac{2}{p_n^- - p_n^+} 
 \begin{bmatrix}
   p_n^- \\ -p_n^+ 
 \end{bmatrix} \;\cdot\; \int_0^1 f(x) \, \sin n\pi x \,dx 
\end{align*}

 &U_n(0)=0 \\ 
 &\Downarrow \\

\end{align*}

 &\Downarrow \\ 
 u(r,\phi) &= \frac{1}{r} + \frac{1}{r^2} \cos\phi

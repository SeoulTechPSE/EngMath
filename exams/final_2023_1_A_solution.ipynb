**1[25].** The equation

> $\displaystyle (1 - x^2)y'' -xy' +\alpha^2 y=0,$

   where $\alpha$ is a constant, is called the *Chebyshev* equation.
  
  (a)$~$ Compute two linearly independent series solutions for $|x|<1$
  
  (b)$~$ Show that for every non-negative integer $\alpha=n$ there is a polynomial solution of degree $n$. When appropriately normalized these are called the *Chebyshev polynomials*
$~$

**Solution**

(a)

> $\begin{align*}
 &\Downarrow \;\;y=\sum_{k=0}^\infty c_n x^k, \;\;x=0 \text{ is an ordinary point}\\ 
 \sum_{k=2}^\infty &c_k k(k-1)x^{k-2}-\sum_{k=2}^\infty c_k k(k-1)x^{k} -\sum_{k=1}^\infty c_k k x^{k} +\alpha^2\sum_{k=0}^\infty c_k x^k =0 \\ 
 2c_2 &+ \alpha^2 c_0 +\left(6c_3 +(\alpha^2-1)c_1 \right) x +\sum_{k=2}^\infty
 \left[(k+2)(k+1) c_{k+2} -(k^2 -\alpha^2)c_k \right] x^k = 0\\ 
 &\Downarrow \\ 
 c_2&=-\frac{\alpha^2}{2} c_0, \;\; c_3=-\frac{\alpha^2 - 1}{6} c_1, \;\;c_{k+2} =-\frac{\alpha^2 -k^2}{(k+2)(k+1)} c_k \\
 &\Downarrow \\
 c_{2k} &= (-1)^k \frac{\left(\alpha^2 - (2k-2)^2\right) \left(\alpha^2 - (2k-4)^2\right) \cdots \left(\alpha^2 - 2^2\right) \alpha^2}{(2k)!} c_0 = \hat{c}_{2k} c_0 \\
 c_{2k+1} &= (-1)^k \frac{\left(\alpha^2 - (2k-1)^2\right) \left(\alpha^2 - (2k-3)^2\right) \cdots \left(\alpha^2 - 1\right)}{(2k+1)!} c_1 =\hat{c}_{2k+1} c_1 \\
 &\Downarrow \\
 y &= c_0 \sum_{k=1}^\infty \hat{c}_{2k} x^{2k} +c_1 \sum_{k=1}^\infty \hat{c}_{2k+1} x^{2k+1} = c_0 y_0(x) + c_1 y_1(x)
\end{align*}$

(b) $~$If $n$ is an even integer, the first series terminates, whereas $y_1(x)$ is an infinite series. For example, if $n=4$, then

> $\displaystyle y_0(x) = c_0 \left[ 1 - \frac{4^2}{2!}x^2 + \frac{(4^2 - 2^2) 4^2}{4!} x^4 \right] =c_0 \left[1 -8x^2 +8x^4 \right]$

  Similarly, when $n$ is an odd integer, the series for $y_1(x)$ terminates with $x^n$. For exampe, if $n=5$, then
  
> $\displaystyle y_1(x) = c_1 \left[ x - \frac{(5^2 -1)}{3!}x^3 + \frac{(5^2 - 3^2)(5^2 -1)}{5!} x^5 \right] =c_1 \left[x -4x^3 +\frac{16}{5}x^5 \right]$
  
  That is, when $n$ is a nonnegative integer, we obtain an $n$-th degree polynomial solution of *Chebyshev* equation.
  
$~$

**2[15].** $~$ Find the general solution of the differential equation on the interval $(0, \infty)$

> $\displaystyle x y'' +y=0$ 

Hint) $~$Use the substitution $\displaystyle y(x)=\sqrt{x}u(x)$, $~2\sqrt{x}=z$
$~$

$~$

**Solution**

By using  $\displaystyle y(x)=\sqrt{x}u(x)~$ and $~2\sqrt{x}=z$, the differential equation becomes

>$\displaystyle \begin{align*}
 y' &=\frac{1}{2\sqrt{x}}u +\sqrt{x}u'\\
 y''&=-\frac{1}{4x\sqrt{x}} u +\frac{1}{\sqrt{x}} u' +\sqrt{x}u''\\
 &\Downarrow \\
 x\sqrt{x} u'' &+\sqrt{x} u' +\left( \sqrt{x} - \frac{1}{4\sqrt{x}} \right )u = 0 \\ 
 u' &=\frac{du}{dz}\frac{dz}{dx}=\frac{1}{\sqrt{x}} \frac{du}{dz} \\
 u'' &=\frac{d}{dz}\left( \frac{du}{dx}\right) \frac{dz}{dx} =\frac{1}{x}\frac{d^2u}{dz^2}-\frac{1}{2x\sqrt{x}}\frac{du}{dz}\\ 
 &\Downarrow \\
 z^2 \frac{d^2u}{dz^2} + z \frac{du}{dz}&+ ( z^2-1)u = 0 
\end{align*}$

The general solution of the above differential equation on the interval $(0, \infty)$ is

>$\displaystyle u(z)=c_1 J_1(z) + c_2 Y_1 (z)$

Thus the general solution on the interval $(0, \infty)$ is

>$\displaystyle y(x)=c_1 \sqrt{x} J_1 (2\sqrt{x}) + c_2 \sqrt{x} Y_1(2\sqrt{x})$
$~$

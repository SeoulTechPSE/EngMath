**1[10].** $~$Solve

> $\displaystyle \frac{dy}{dx}=\frac{xy+3x-y-3}{xy -2x +4y-8}$

$~$**solution**

$$
\begin{align*}
\frac{dy}{dx} &= \frac{(y+3)(x-1)}{(y-2)(x+4)} \\ 
 &\Downarrow \\ 
 \left ( 1-\frac{5}{y+3} \right )\,dy &= \left ( 1-\frac{5}{x+4} \right )\, dx\\
 \\ 
 y -5\ln \left| y+3 \right| &= x -5\ln \left| x+4 \right| +C\\
 \\ 
 x -y + 5 \ln &\left| \frac{y+3}{x+4}\right| = C 
\end{align*}
$$

**3[10].** $~$ Solve $\cos x dx + \left( 1+\frac{2}{y} \right)\sin x dy=0$

$~$
**solution**

\begin{align*}
 &\Downarrow \times \;\; y^2 e^y \\
 y^2e^y \cos x \,dx &+ y^2 e^y \left ( 1 + \frac{2}{y} \right ) \sin x \, dy = 0 \\ 
 &\Downarrow \\ 
 \frac{\partial f}{\partial x} = y^2 e^y \cos x &\;\;\Rightarrow\;\; f = y^2e^y \sin x +h(y) \\ 
 \frac{\partial f}{\partial y} = y^2 e^y \left ( 1 + \frac{2}{y} \right ) \sin x +\frac{dh}{dy}\;\; &\Rightarrow \;\;\frac{dh}{dy} = 0\\
 &\Downarrow \\
 f &=  y^2e^y \sin x +C = 0
\end{align*}

**4[10].** $~$Solve $~\displaystyle \frac{dy}{dx} = \frac{x}{y} + \frac{y}{x} +1$

$~$

\begin{align*}
 &\Downarrow \;\;y=ux, \;\;\frac{dy}{dx}=u+x\frac{du}{dx} \\ 
 u + x\frac{du}{dx} &= \frac{1}{u} + u + 1 \\ 
 \left ( 1 - \frac{1}{1+u} \right ) du &= \frac{1}{x} dx\\ 
 &\Downarrow \\ 
 u - \ln |1 + u| &= \ln |x| +C \\
 &\Downarrow \\
 \frac{y}{x} -\ln |x &+ y| = C 
\end{align*}

**5[15].** $~$Solve $~y'' - 2y' + y = 4x^2 - 3 +x^{-1} e^x$

$~$
**solution**

$$
\begin{align*}
  y_h'' -2y_h' +y_h &= 0, \;\;y_{h_1} = e^x, \;y_{h_2}=xe^x \\ \\
  y_{p_1}'' -2y_{p_1}' +y_{p_1} &= 4x^2-3\\ 
  &\Downarrow\, y_{p_1} = Ax^2 + Bx + C \\
 y_{p_1} &= 2x^2 +8x +9 \\ \\
 y_{p_2}'' -2y_{p_2}' +y_{p_2} &= x^{-1} e^x \\
 W &= \begin{vmatrix}
       e^x & xe^x\\ 
       e^x & (1+x)e^x
      \end{vmatrix} = e^{2x} \\
 W_1 &= \begin{vmatrix}
         0 & xe^x\\ 
         x^{-1}e^x & (1+x)e^x \\
        \end{vmatrix} = -e^{2x} \\
 W_2 &= \begin{vmatrix}
          e^x & 0 \\ 
          e^x & x^{-1}e^x \\
        \end{vmatrix} = x^{-1} e^{2x} \\
u_1 = \int \frac{W_1}{W} \,dx = -x, &\;\;u_2 = \int \frac{W_2}{W}\,dx = \ln x \\
&\Downarrow \\
y_{p_2} &=-xe^x + xe^x \ln x
\end{align*}$$

Then

$$
y = c_1e^x + c_2 xe^x +2x^2 +8x +9 -xe^x + +xe^x \ln x
$$

**6[15].** Solve $x^2y'' -4xy' + 6y = 2x^4 + x^2$

$~$
**solution**

\begin{align*}
x^2 y_h'' -4xy_h' +6y_h &= 0, \;\;y_{h_1} = x^2, \;y_{h_2}=x^3 \\ \\
 y_{p}'' -\frac{4}{x}y_{p}' +\frac{6}{x^2}y_{p} &= 2x^2 +1 \\
 W &= \begin{vmatrix}
       x^2 & x^3\\ 
       2x & 3x^2
      \end{vmatrix} = x^4 \\
 W_1 &= \begin{vmatrix}
          0 & x^3\\ 
          2x^2+1 & 3x^2 \\
        \end{vmatrix} = -2x^5 -x^3 \\
 W_2 &= \begin{vmatrix}
          x^2 & 0 \\ 
          2x & 2x^2+1 \\
        \end{vmatrix} = 2x^4+x^2 \\
u_1 = \int \frac{W_1}{W} \,dx = -x^2 -\ln x, &\;\;u_2 = \int \frac{W_2}{W}\,dx = 2x -\frac{1}{x} \\
&\Downarrow \\
y_{p} &= x^4 -x^2 -x^2 \ln x
\end{align*}

Then

$$y = c_1x^2 +c_2 x^3 +x^4 -x^2 -x^2 \ln x$$

**8[15].** $~$ One definition of the gamma function $\Gamma(\alpha)$ is given by the improper integral

  >$\displaystyle \Gamma(\alpha) =\int_0^\infty t^{\alpha-1} e^{-t}\,dt, \;\;\alpha > 0$

  > Use this definition to show that $\Gamma(\alpha + 1)=\alpha \Gamma(\alpha)$ and show that
  
  >$\displaystyle \mathcal{L} \{ t^\alpha \} = \frac{\Gamma (\alpha+1)}{s^{\alpha+1}}, \;\; \alpha>-1$
  
$~$
**solution**

\begin{align*}
 \Gamma(\alpha+1) &= \int_0^\infty t^\alpha e^{-t} \,dt \\ 
 &= \left. -t^\alpha e^{-t} \right|_0^\infty + \alpha\int_0^\infty t^{\alpha-1} e^{-t}\,dt \\ 
 &= \alpha \Gamma(\alpha)\\ \\
 \mathcal{L}[t^\alpha] &= \int_0^\infty t^\alpha e^{-st} \,dt \\
 &\Downarrow \,st=t', \;\;dt=\frac{1}{s}dt' \\
 &=\int_0^\infty \left( \frac{t'}{s} \right )^\alpha e^{-t'} \,\frac{1}{s} \,dt' \\
 &= \frac{\Gamma (\alpha+1)}{s^{\alpha+1}} 
\end{align*}

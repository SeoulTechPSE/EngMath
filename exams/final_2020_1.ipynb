**2.** $\text{ }$ Find the general solution of differential equation

> $y'' -2y'+y =x^2 e^x$

> **solution**

> Applying $(D−1)^3$ to the differential equation, we obtain $(D−1)^3(D^2−2D+1) = (D−1)^5 = 0$.
  Then
  
> $y = \underbrace{c_1 e^x + c_2 x e^x}_{y_h} +\underbrace{A x^2 e^x +B x^3 e^x +C x^4 e^x}_{y_p}$

> Substituting $y_p$ into the differential equation yields

> $2A e^x + 6B xe^x +12C x^2e^x = x^2 e^x$

> Equating coefficients gives $A = 0$, $B = 0$, and $C = 1/12$. The general solution is

> $y = c_1 e^x +c_2 xe^x +\frac{1}{12} x^4e^x$

**3.** $\text{ }$ Find a Cauchy-Euler differential equation $ax^2y'' +bxy' +cy=0$, where $a$, $b$, and $c$ are real constants, if it is known that

> (a) $m_1=3$ and $m_2=-1$ are roots of its auxiliary equation,

> (b) $m_1=i$ is a complex root of its auxiliary equation

> **solution**

> (a) The auxiliary equation is $am(m−1)+bm+c = am^2 +(b−a)m+c = 0$. If the roots are
$3$ and $−1$, then we want $(m − 3)(m + 1) = m^2 − 2m − 3 = 0$. Thus, let $a = 1$, $b = −1$,
and $c = −3$, so that the differential equation is $x^2y'' −xy' −3y = 0$

> (b) In this case, we want the auxiliary equation to be $m^2 + 1 = 0$, so let $a = 1$, $b = 1$, and
$c = 1$. Then the differential equation is $x^2y'' +xy' +y = 0$

**7.** Use an appropriate infinite series method about $x=0$ to find two solutions of the given differential equation

> $xy'' -(x+2)y' +2y=0$

> **solution**

> Substituting $y = \sum_{n=0}^\infty c_n x^{n+r}$ into the differential equation, we obtain

> $x y'' -(x +2)y'+2y = (r^2 -3r) c_0 x^{r -1}
+\sum_{k=1}^\infty \left[(k + r)(k + r -3)c_k -(k +r -3) c_{k-1} \right] x^{k +r -1} = 0$

> which implies

> $r^2 -3r = r(r − 3) = 0\text{ }$ and $\text{ }(k + r)(k + r − 3)c_k − (k +r −3)c_{k−1} = 0$

> The indicial roots are $r_1 = 3$ and $r_2 = 0$. For $r_2 = 0$, $\text{ }$ the recurrence relation is

> $k(k − 3)c_k − (k −3)c_{k−1} = 0, \text{ } k = 1, 2, 3, \cdots$

> Then $c_1 −c_0 = 0$, $2c_2 −c_1 = 0$, $0c_3 − 0c_2 = 0$. Therefore $c_3$ is arbitrary and

> $c_k = \frac{1}{k}c_{k−1}, \text{ }k = 4, 5, 6, \cdots$

> Taking $c_0 \neq 0$ and $c_3 = 0$, we obtain

> $c_1 = c_0$, $\text{ }c_2 = \frac{1}{2} c_0$, $\text{ }c_3 = c_4 = c_5 = \cdots = 0$

> Taking $c_0 = 0$ and $c_3 \neq 0$, we obtain

> $c_0 = c_1 = c_2 = 0$, $\text{ }c_4 =\frac{1}{4} c_3 = \frac{6}{4!} c_3$, 
  $\text{ }c_5 = \frac{1}{5 \cdot 4} c_3 = \frac{6}{5!} c_3$, $\text{ }c_6 = \frac{1}{6\cdot 5\cdot 4} c_3 = \frac{6}{6!} c_3$,
  
> and so on. In this case we obtain the two solutions

> $y_1 = 1 +x +\frac{1}{2} x^2$

> and

> $y_2 = x^3 +\frac{6}{4!} x^4 +\frac{6}{5!} x^5 +\frac{6}{6!} x^6 +\cdots = 6\left[ e^x −\left(1 +x +\frac{1}{2} x^2\right)\right]$

**9.** $\text{ }$ Identify the given conic section and graph

> $16x^2 +24xy +9y^2 -3x +4y=0$

> **solution**

> The given equation can be written as

> $\begin{pmatrix}
x & y
\end{pmatrix} 
\begin{pmatrix}
16 & 12\\ 
12 & 9
\end{pmatrix}
\begin{pmatrix}
x \\ y
\end{pmatrix} +
\begin{pmatrix}
-3 & 4 
\end{pmatrix}
\begin{pmatrix}
x \\ y
\end{pmatrix} = 0
$

> Using $\lambda_1 = 25$, $\text{ }\lambda_2 = 0$, 
$\text{ }\mathbf{k}_1 = \begin{pmatrix}
4 \\ 
3
\end{pmatrix}$, $\text{ }\mathbf{k}_2 = \begin{pmatrix}
\;\;\,3 \\ 
-4
\end{pmatrix}$,
$\mathbf{P} = \frac{1}{5}
\begin{pmatrix}
 4& \;\;\,3 \\ 
 3& -4
\end{pmatrix}$ and $\mathbf{x} =\mathbf{P}\mathbf{\hat{x}}$, $\text{ }$ we find

> $\begin{pmatrix}
\hat{x} & \hat{y}
\end{pmatrix} 
\begin{pmatrix}
25 & 0\\ 
0 & 0
\end{pmatrix}
\begin{pmatrix}
\hat{x} \\ \hat{y}
\end{pmatrix} -
\begin{pmatrix}
0 & 5 
\end{pmatrix}
\begin{pmatrix}
\hat{x} \\ \hat{y}
\end{pmatrix} = 0
$

> The conic section is a parabola $\hat{y} = 5 \hat{x}^2$. Now from $\mathbf{\hat{x}} = \mathbf{P}^T\mathbf{x}$, we see that the $\hat{x}\hat{y}$-coordinates of $(0, 5)$ and $(5, 0)$ are $(3,−4)$ and $(4, 3)$, respectively. From this we conclude that the $\hat{x}$-axis and $\hat{y}$-axis are as shown in the accompanying figure.

> **solution**

> We have

> $\mathbf{x}_h = c_1 \begin{pmatrix}

\;\;\, 1 \\ -1
\end{pmatrix} t e^{2t} +\begin{pmatrix}

\end{pmatrix}  e^{2t}\right]$

> $\Phi = 
\begin{pmatrix}

\begin{pmatrix}
 -te^{-2t} & -te^{-2t} -e^{-2t}\\ 
 \;\;e^{-2t} & e^{-2t}

> $\displaystyle \mathbf{u} = \int \Phi^{-1} \mathbf{f} \,dt

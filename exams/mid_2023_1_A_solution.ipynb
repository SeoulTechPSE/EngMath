**1[20].** $~$Find a general solution of the equation $~\displaystyle ydx+(2x+xy-3)dy=0$


$~$

* **Solution**

\begin{align*}
 ydx +(2x+xy-3)&dy =0 \\ 
 &\Downarrow \\ 
 \frac{dx}{dy} + \left(1 + \frac{2}{y} \right)x &= \frac{3}{y} \\
 &\Downarrow \\  
 \frac{d}{dy} \left[ x \,y^2 e^y \right ] &= 3y e^y \\
 &\Downarrow \\ 
 \color{red}{x = \frac{3}{y}\left(1-\frac{1}{y} \right)}\, & \color{red}{\, + \,Cy^{-2} e^{-y}}
\end{align*}

$$~\text{or}~$$

\begin{align*}
 ydx \;+\;&(2x+xy-3)dy =0 \\ 
 &\Downarrow \;\; \times \, y e^y\\ 
 y^2 e^y dx \;+\;&y e^y (2x+xy-3)dy =0 \\ 
 &\Downarrow \\  
 \frac{\partial f}{\partial x} = y^2 e^y \;\;&\rightarrow \;\; f(x,y) =y^2 e^y x + h(y)\\
 &\Downarrow \\
 \frac{\partial f}{\partial y} = ye^y(2x +xy) +\frac{dh}{dy} \;\;&\rightarrow \;\; \frac{dh}{dy} = -3y e^y\\
 &\Downarrow \\  
 h(y) =-3e^y (y-1) - C \;\;&\rightarrow \;\; f(x,y) =y^2 e^y x -3e^y (y-1) = C \\ 
 &\Downarrow \\ 
 \color{red}{x = \frac{3}{y}\left(1-\frac{1}{y} \right)}\, & \color{red}{\, + \,Cy^{-2} e^{-y}}
\end{align*}

**2[15].** $~$Solve the differential equation $~\displaystyle xv\frac{dv}{dx} +v^2=32x$


$~$

* **Solution**

\begin{align*}
 xv\frac{dv}{dx} +v^2 &=32x \\ 
 &\Downarrow \\
 \frac{dv}{dx} +\frac{1}{x} v &= \frac{32}{v}\\
 &\Downarrow \;\; u = v^2, \; \frac{du}{dx} = 2v \frac{dv}{dx}\\ 
 \frac{du}{dx} +\frac{2}{x}u &= 64\\ 
 &\Downarrow \\
 \frac{d}{dx} \left( u x^2 \right ) &= 64x^2 \\
 &\Downarrow \\
 u = \frac{64}{3} x + \frac{C}{x^2} \;\;&\rightarrow \;\; \color{red}{v = \pm\sqrt{\frac{64}{3} x + \frac{C}{x^2}}}
\end{align*}

**3[15].** $~$Solve $~y'' - 2y' +y = 4x^2 - 3 +x^{-1} e^x$


$~$

* **Solution**

1. Homogenous part

\begin{align*}
 y'' -& 2y' + y =0 \\ 
 &\Downarrow \;\; y=e^{px} \\ 
 p^2 - 2p + 1 = 0 \;\; &\rightarrow p = 1 \\ 
 &\Downarrow \\ 
 y_1 = e^x, &\;\; y_2 = x e^x
\end{align*}

$~$

2. Nonhomogeneous part: Method of undetermined coefficient

\begin{align*}
 y_{p_1}'' -& 2y_{p_1}' + y_{p_1} = 4x^2 -3 \\ 
 &\Downarrow \;\; y_{p_1}=Ax^2 +Bx +C \\ 
 2A -4Ax -2B +Ax^2 +Bx + C &=4x^2 -3   \\ 
 &\Downarrow \\ 
 A=4, \; B=16, \; C=21 \;\;&\rightarrow \;\; y_{p_1} = 4x^2 +16x +21
\end{align*}

$~$

3. Nonhomogeneous part: Method of variation of parameters

\begin{align*}
 y_{p_2}'' -& 2y_{p_2}' + y_{p_2} = x^{-1} e^x \\ 
 &\Downarrow \\ 
 W = \begin{vmatrix}
  e^x & x e^x\\ 
  e^x & (x+1) e^x
 \end{vmatrix} = e^{2x}, \;\;
 W_1 = &\begin{vmatrix}
  0 & x e^x\\ 
  x^{-1}e^x & (x+1) e^x
 \end{vmatrix} = -e^{2x}, \;\;
 W_2 = \begin{vmatrix}
  e^x & 0 \\ 
  e^x & x^{-1}e^x
 \end{vmatrix} = x^{-1}e^{2x} \\
 &\Downarrow \\
 u_1 &= \int \frac{W_1}{W} \,dx = -x \\
 u_2 &= \int \frac{W_2}{W} \,dx = \ln x \\
 &\Downarrow \\
 y_{p_2} &= -x e^x +x e^x \ln x
\end{align*}
$~$

$$\color{red}{y = c_1 e^x + c_2 x e^x +4x^2 +16x +21+x e^x \ln x}$$

**4[15].** $~$Solve the given initial-value problem on the interval $(-\infty, 0)$.

$\;\;\;\;\;\;\;\;\;4x^2y'' + y = 0, \;y(-1)=2, \;y'(-1)=4$


$~$

* **Solution**

\begin{align*}
 4x^2y'' &+ y = 0, \;\;y(-1)=2, \;y'(-1)=4 \\ 
 &\Downarrow \;\; x=-t \\ 
 4 t^2 \frac{d^2y}{dt^2} + y &= 0, \;\;y(1)=2, \;y'(1)=4\\ 
 &\Downarrow \;\; y=t^p\\
 4p(p-1) +1 = 0 \;\;&=\rightarrow \;\; p=\frac{1}{2}\\
 &\Downarrow\\
 y =c_1 t^{1/2} + c_2 t^{1/2}\ln t \;\;&\rightarrow \;\;c_1=2, \;\;c_2=3\\
 &\Downarrow\\
 \color{red}{y = 2 \sqrt{-x}}\; &\; \color{red}{\;+ 3\sqrt{-x} \ln(-x)}
\end{align*}

**5[20].** $~$ $~y=e^{-t^2}$ is a solution of the initial value problem 
$~\displaystyle\frac{dy}{dt} +2ty=0, \;y(0)=1$. $~$Find $~Y(s)=\mathcal{L}\left[ e^{-t^2}\right]$ by applying
the Laplace transform to the given differential equation


$\,\;\;\;\;\;\;\;\;$Hint) $\displaystyle\int_0^\infty e^{-t^2}\, dt = \frac{\sqrt{\pi}}{2}, \;\;\mathrm{erf}(t) = \frac{2}{\sqrt{\pi}} \int_0^t e^{-u^2}\,du, \;\;\mathrm{erf}c(t) = 1 - \mathrm{erf}(t)$

$~$

* **Solution**

\begin{align*}
 \frac{dy}{dt} +2ty&=0, \;y(0)=1 \\ 
 &\Downarrow \\ 
 sY -1 -2\frac{dY}{ds} = 0, \;\;&Y(0)=\int_0^\infty e^{-t^2}\,dt = \frac{\sqrt{\pi}}{2} \\ 
 &\Downarrow \\ 
 \frac{d}{ds} \left[Y\cdot\exp\left(-\frac{1}{4}s^2 \right) \right] &= -\frac{1}{2} \exp \left(-\frac{1}{4}s^2 \right) \\ 
 &\Downarrow \\
 Y(s) = \frac{\sqrt{\pi}}{2} \exp\left(\frac{1}{4}s^2 \right) -&\frac{1}{2}\exp\left(\frac{1}{4}s^2 \right) \int_0^s \exp\left(-\frac{1}{4}s^2 \right) \,ds \\
 &\Downarrow \frac{1}{2}s=u, \;\frac{1}{2}ds=du \\
 \color{red}{Y(s) =} \frac{\sqrt{\pi}}{2} \exp\left(u^2\right) \left[ 1 - \frac{2}{\sqrt{\pi}} \int_0^{\frac{1}{2}s} \exp\left(-u^2\right)\, du \right]
    &\color{red}{\;= \frac{\sqrt{\pi}}{2} \exp\left(\frac{1}{4}s^2 \right)  \mathrm{erfc}\left(\frac{1}{2}s \right)}
\end{align*}

**6[15].** Solve the given differential equation $\displaystyle ~ y''+2y'+y=(t-1) u(t-1) +\delta(t-5)$, $y(0)=0$, $y'(0)=0$

$~$

* **Solution**

\begin{align*}
 y''+2y'+y=(t-1) u(t-1) &+\delta(t-5), \;\;y(0)=0,\; y'(0)=0 \\ 
 &\Downarrow \\ 
 (s^2 +2s +1)\,Y &= \frac{e^{-s}}{s} + e^{-5s}\\ 
 &\Downarrow \\ 
 Y = \frac{e^{-s}}{s(s+1)^2} + \frac{e^{-5s}}{(s+1)^2} &= \left[ \frac{1}{s} -\frac{1}{s+1} -\frac{1}{(s+1)^2}\right] e^{-s} + \frac{e^{-5s}}{(s+1)^2} \\
 &\Downarrow \\
 \color{red}{y(t) = \left[t-1 -e^{-(t-1)} -(t-1)e^{-(t-1)} \right]} \; & \color{red}{\;u(t -1) +(t-5) e^{-(t-5)} u(t-5)}
\end{align*}
